## Sesión 8. Plotly

```python
import plotly.express as px 
import pandas as pd 
```

```python
gapminder = px.data.gapminder() 
```

```python
gapminder.head() 
```

```python
gapminder["country"].unique()
```

```python
gapminder["year"].describe()
```

```python
# Seleccionar un país
pais = gapminder.loc[gapminder["country"] == 'Peru', :].reset_index(drop = True)
pais
```

```python
# Gráfico de barras
fig = px.bar(pais, x = 'year', y = 'pop')
fig.show() 
```

```python
fig = px.bar(pais, x = 'year', y = 'pop', color_discrete_sequence = ['darkorange'])
fig.show() 
```

```python
fig = px.bar(pais, x = 'year', y = 'pop', color = 'lifeExp', labels = {'pop': 'Población'})
fig.show() 
```

[Paletas de colores de Plotly](https://plotly.com/python/builtin-colorscales/)

```python
px.colors.named_colorscales()
```

```python
fig = px.bar(pais, x = 'year', y = 'pop', color = 'lifeExp', labels = {'pop': 'Población'},
             color_continuous_scale = "YlGn")
fig.show() 
```

```python
continent_pop = gapminder.groupby(["continent", "year"])["pop"].sum().reset_index()
```

```python
fig = px.bar(continent_pop, x = 'year', y = 'pop', color = 'continent', 
             labels = {'pop': 'Población', "year": "Año", "continent": "Continente"},
             color_discrete_sequence = ["#e76f51", "#f4a261", "#e9c46a", "#2a9d8f", "#264653"])
fig.show() 
```

```python
gapminder2007 = gapminder.loc[gapminder["year"] == 2007, ].reset_index(drop = True)
gapminder2007
```

```python
fig = px.scatter(gapminder2007, x='gdpPercap', y='lifeExp')
fig.show() 
```

```python
fig = px.scatter(gapminder2007, x='gdpPercap', y='lifeExp', color='continent')
fig.show() 
```

```python
fig = px.scatter(gapminder2007, x='gdpPercap', y='lifeExp', color='continent', size='pop', size_max=60, 
    height = 500, log_x = True)
fig.show() 
```

```python
fig = px.scatter(gapminder2007, x='gdpPercap', y='lifeExp', color='continent', size='pop', size_max=60, 
    height = 500, log_x = True, hover_name = 'country')
fig.show() 
```

```python
fig = px.scatter(gapminder, x='gdpPercap', y='lifeExp', color='continent', size='pop', size_max=60, 
                 height = 500, log_x = True, hover_name = 'country',
                 animation_frame='year', animation_group='country')
fig.show()
```

```python
fig = px.scatter(gapminder2007, x='gdpPercap', y='lifeExp', color='continent', size='pop', size_max=60, 
                 hover_name='country', facet_row='continent', height = 800)
fig.show() 
```

```python
fig = px.scatter_geo(gapminder2007, locations = 'iso_alpha', color='continent', projection='orthographic',
                     size = "pop",
                     hover_data = {'iso_alpha': False, 'country': True, 'gdpPercap': True, 'lifeExp': True})
fig.show() 
```

```python
fig = px.choropleth(gapminder, locations='iso_alpha', color='lifeExp', hover_name='country', 
                    animation_frame='year', color_continuous_scale="RdYlGn", 
                    projection='natural earth')
fig.show() 
```

```python
df = px.data.tips()

fig = px.density_heatmap(df, x="total_bill", y="tip", nbinsx=10, nbinsy=10, color_continuous_scale="Viridis")
fig.show()
```

```python
fig = px.density_heatmap(df, x="total_bill", y="tip", 
                         marginal_x="histogram", marginal_y="histogram",)
fig.show()
```

```python
fig = px.histogram(df, x="tip", color="sex")
fig.show()
```

```python
fig = px.histogram(df, x="tip", color="sex", marginal="rug",
                   hover_data=df.columns)
fig.show()
```

```python
fig = px.histogram(df, x="tip", color="sex",
                   marginal="box", # or violin, rug
                   hover_data=df.columns)
fig.show()
```

```python
df = gapminder.query("country in ['Canada', 'Botswana']")

fig = px.line(df, x="lifeExp", y="gdpPercap", color="country", text="year")
fig.update_traces(textposition="bottom right")
fig.show()
```
