```python
import requests
from bs4 import BeautifulSoup
import pandas as pd

# Hacer la solicitud HTTP
url = "https://es.wikipedia.org/wiki/Anexo:%C3%93scar_a_la_mejor_pel%C3%ADcula"
response = requests.get(url)

# Parsear el HTML usando BeautifulSoup
soup = BeautifulSoup(response.text, 'html.parser')

# Encontrar la tabla por clase
table = soup.find_all('table')[1]

# Extraer los datos de la tabla y convertirlos en una lista de listas
data = []
rows = table.find_all('tr')
for row in rows:
    cols = row.find_all(['th', 'td'])
    cols = [col.text.strip() for col in cols]
    data.append(cols)

# Convertir la lista de listas en un DataFrame de Pandas
df = pd.DataFrame(data)

# Mostrar el DataFrame
df
```
