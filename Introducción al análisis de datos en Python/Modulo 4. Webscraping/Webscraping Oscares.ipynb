```python
import requests
from bs4 import BeautifulSoup
import pandas as pd

# Hacer la solicitud HTTP
url = "https://es.wikipedia.org/wiki/Anexo:%C3%93scar_a_la_mejor_pel%C3%ADcula"
response = requests.get(url)

# Parsear el HTML usando BeautifulSoup
soup = BeautifulSoup(response.text, 'html.parser')

# Extraer los datos de la tabla y convertirlos en una lista de listas
data = []

for i in range(1, 12):
    # Encontrar la tabla por clase
    table = soup.find_all("table")[i]
    filas = table.find_all("tr")

    for fila_i in filas:
        cols_i = fila_i.find_all(["th", "td"])
        cols_i = [col.text.strip() for col in cols_i]
        data.append(cols_i)

# Convertir la lista de listas en un DataFrame de Pandas
df = pd.DataFrame(data)

# Mostrar el DataFrame
df
```

```python
df.columns = df.iloc[0,:].values
df = df.loc[df.iloc[:, 0] != "Título original y en español", :].reset_index(drop = True)
df
```

```python
filtro = df.iloc[:, 0].str.contains("ceremonia de entrega de los Premios")
ceremonia = df.loc[filtro,].iloc[:, 0].values
df = df.loc[~filtro,:].assign(Ceremonia = ceremonia).reset_index(drop = True)
df
```
