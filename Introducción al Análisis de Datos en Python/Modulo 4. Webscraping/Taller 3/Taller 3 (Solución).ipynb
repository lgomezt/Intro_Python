# 1. Scraping Win Sports
Ingrese a la página web [www.winsports.co](https://www.winsports.co/) y realice un código en Python que haga un webscraping de la tabla de posiciones de la **Liga BetPlay Dimayor 2024-I**. El resultado final debe ser un dataframe

```python
import requests
import pandas as pd

r = requests.get("https://files-optafeeds-produccion-1.s3.amazonaws.com/positions/371/2024/all.json")
diccionario = r.json()

equipos = diccionario["phases"]['1185']["teams"]

consolidado = pd.DataFrame()
for nombre, fila in zip(equipos.keys(), equipos.values()):
    temp = pd.DataFrame(fila, index = [nombre])
    consolidado = pd.concat([consolidado, temp], axis = 0)
consolidado = consolidado.reset_index(drop = True)
consolidado
```

### 2.2. Descargue la información.
Haga otro webscraping que vaya iterando sobre cada uno de los elementos de la lista de inmuebles y extraiga la información relevante del inmueble. Construya un dataframe con sus resultados.

```python
from bs4 import BeautifulSoup
import json
from tqdm import tqdm

inmuebles = pd.DataFrame()
for enlace in tqdm(enlaces):
    r = requests.get("https://www.metrocuadrado.com" + enlace)
    if r.status_code == 200:
        soup = BeautifulSoup(r.content)
        data = soup.find_all(id = "__NEXT_DATA__")[0]
        data_dict = json.loads(data.contents[0])
        fila = data_dict['props']['initialState']['realestate']["basic"]
        fila_temp = pd.DataFrame([fila])
        inmuebles = pd.concat([inmuebles, fila_temp], axis = 0)
    else:
        print(f"Revisar {enlace}")
inmuebles = inmuebles.reset_index(drop = True)        
```
